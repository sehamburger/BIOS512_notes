```r
library('tidyverse')
library('palmerpenguins')
```

```r
penguins2 = penguins |> mutate(species = species |> factor(levels = c('Adelie', 'Gentoo', 'Chinstrap'))) #manually setting the unique orders of values in factor

penguins2 |> head()
```

```r
p = ggplot(penguins2, aes(y = species)) + geom_bar()

p
```

```r
penguins3 = penguins |> 
    mutate(species = species |> fct_infreq() |> fct_rev())

penguins3 |> pull() |> levels()
```

```r
p = ggplot(penguins3, aes(y = species)) + geom_bar()

p
```
