```r
library('tidyverse')
library('rvest')
library('lubridate')
options(repr.plot.res = 300, repr.plot.width = 6, repr.plot.height = 6)
```

```r
scrape_table = function(i, MF='Women\'s') {
    url = "https://www.baa.org/races/boston-marathon/results/champions"
    webpage = xml2::read_html(url)
    tbl = html_nodes(webpage, "table")[[i]] %>%
        html_table() %>%
        drop_na() %>%
        magrittr::set_colnames(c('year', 'name', 'country', 'time')) %>%
        mutate(time = hms(time)) %>%
        mutate(race.time = as.numeric(time, 'hours')) %>%
        mutate(MF = MF) %>%
        mutate(
            country.simple = ifelse(
                country %in% c('United States', 'Ethiopia', 'Japan', 'Kenya'),
                country,
                'other'
            )
        )
    tbl
}

df.men = scrape_table(1, MF='Men\'s')
df.women = scrape_table(2)

df.marathon = rbind(df.men, df.women)
```

<span style  = "color:blue;"> When we define an encoding, it's what ggplot2 defines a scale
- can be an axis, a color palette; once we've created these scales - we can adjust them using the scale functions

```r
options(repr.plot.resolution = 200, repr.plot.width = 4)

df.marathon = df.marathon |> 
    mutate(year = as.Date(paste(year, 1, 1, sep = "-")))

#sometime in programs - like inscape - you can hover over colors to get hex-formats
#htmlcolorcodes.com
colors = c('#9cca9e', '#ef9285', '#b983b7', '#cecece', '#56b6be') #creating a list - or vector - of values
bg = '#f0f0f0'

m = ggplot(df.marathon, aes(x = year, y = time, color = country.simple)) + 
    geom_point(na.rm = T) +
    facet_wrap(~MF, ncol = 1, scales = 'free_x') +
    
    #adjusting scale
    scale_y_time(labels = scales::label_time(format('%H:%M'))) + #format('%H:%M:%S') will give H:M:Seconds
    scale_x_date(date_breaks = "10 years", date_labels = '%y') +
    
    #providing color palette
    scale_color_manual(values = colors) +

    #adding and changing labels
    labs(x = '', y = 'Winning time', title = "A slower field in 2018's Boston Marathon", 
        subtitle = "Finish time winners of the Boston Marathon, by country") +

    #theme functions - default theme is theme_gray()
    #if there's a pre-set theme you like, you can chagne a lot of different chart elemets at once
 #    theme_bw() +  
 #     theme(legend.position = "none")
 #    # legend.background = element_rect(color = 'blue', fill = 'blue') #color of line outside legend
 #    # plot.background = element_rect(color = 'blue') +
 #    # panel.background = element_rect(color = 'blue') +
 #    #ajusting axis text
 #    axis.text.x = element_text(angle = 45) 
 # #   axis.title.y = element_text(color = 'green')

theme(legend.position = 'None',
          plot.margin = margin(0.5, 0.5, 0.5, 0.5, "cm"),
          plot.subtitle = element_text(color = '#212121', size = 8),
          plot.title = element_text(color = '#242424', face = 'bold'), #appearance of plot title text
          strip.text = element_text(hjust = 0.9, color = '#010101', face = 'bold'),
          plot.title.position = 'plot', #justifies the title on the left side
          plot.background = element_rect(fill = bg),
          panel.background = element_rect(fill = bg),
          strip.background = element_blank(),
          strip.placement = 'inside',
          panel.grid.major = element_line(color = '#e0e0e0'),
          panel.grid.minor = element_blank(),
          panel.spacing = unit(0.5, "cm"),
          axis.ticks = element_blank(),
          axis.title.y = element_text(face = 'bold', margin = margin(r = 10)),
          axis.text = element_text(color = '#9d9d9d', family = 'DecimaMonoPro', size = 8))

#common that we want to label specific datapoints in our chart - this is called annotations
#geom_text() will print values from label encoding onto the chart
#library(ggrepel) -- way to add annotations to charts such that they're not overlapping and arranged in a useful way

m
```
